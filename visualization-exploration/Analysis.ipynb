```python
# import dependencies
import pandas as pd
```

```python
#import dataset

card_data = pd.read_json('all-cards-20250706092749.json')
card_data
```

```python
# check random sample of the dataset
card_data.sample(10)
```

```python
# check datatypes, null values, and memory usage
card_data.info()
```

```python
# import matplotlib for plotting
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick

# set up plotting style
plt.style.use('default')
```

```python
# extract prices from the dataset and check the data
card_prices = card_data[['prices']].copy()
card_prices.head()
```

```python
# Expand the 'prices' dictionary column into separate columns
card_prices = card_prices['prices'].apply(pd.Series)
card_prices
```

```python
# Count how many entries have None (or NaN) in each column
none_counts = card_prices.isna().sum()
print("Number of None/NaN values in each column:")
print(none_counts)

# Count how many rows have all None/NaN values (i.e., only None entries)
only_none_rows = card_prices.isna().all(axis=1).sum()
print(f"\nNumber of rows with only None/NaN values: {only_none_rows}")

card_prices = card_prices.dropna(how='all')
```

```python
# Calculate percentage of None/NaN values for each column in card_prices
percent_none = card_prices.notna().mean() * 100
print("\nPercentage of None/NaN values in each column:")
print(percent_none)


# Plot the actual data
# Plot bars with value 100 for each entry (for reference/comparison)
plt.bar(percent_none.index, [100]*len(percent_none), color='red', alpha=0.4, label='100% Reference', width=0.5, align='center')

# Plot the actual data
ax = percent_none.plot.bar()
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.ylabel('Percentage of valid values')
plt.title('Percentage of valid values per column in card_prices')
plt.show()
```

```python
usd_prices = card_prices['usd'].dropna().astype(float)
usd_prices.sample(10)
```

```python
usd_prices.value_counts().sort_index().plot(kind='line', figsize=(15, 5))
plt.xlabel('USD Price')
plt.ylabel('Count')
plt.title('Distribution of USD Prices')
```

```python
q_high = usd_prices.quantile(0.95)
q_low = usd_prices.quantile(0.05)
usd_prices_filtered = usd_prices[(usd_prices >= q_low) & (usd_prices <= q_high)]
usd_prices_filtered.value_counts().sort_index().plot(kind='line', figsize=(15, 5))
plt.xlabel('USD Price')
plt.ylabel('Count')
plt.title('Distribution of USD Prices (Filtered 5% Quantiles)')
```
